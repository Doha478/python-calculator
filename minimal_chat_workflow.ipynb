```python
from langchain.chat_models import init_chat_model
from langgraph.graph import StateGraph, END
from typing import TypedDict, Optional
import getpass
import os


os.environ["MISTRAL_API_KEY"] = getpass.getpass("Enter API key")

MISTRAL_API_URL = "https://api.mistral.ai/v1/chat/completions"

# SchÃ©ma dâ€™Ã©tat
class EtatIA(TypedDict):
    question: str
    reponse: Optional[str]

# Initialiser le modÃ¨le
model = init_chat_model("mistral-large-latest", model_provider="mistralai")

# NÅ“ud du graphe : rÃ©ponse Ã  une question
def noeud_reponse(state: EtatIA) -> EtatIA:
    question = state["question"]
    result = model.invoke(question)  # PAS besoin de HumanMessage ici
    return {
        "question": question,
        "reponse": result.content
    }

# Construction du graphe
graph = StateGraph(EtatIA)
graph.add_node("repondre", noeud_reponse)
graph.set_entry_point("repondre")
graph.add_edge("repondre", END)
app = graph.compile()

# ExÃ©cution du graphe
etat_initial: EtatIA = {
    "question": "Bonjour, peux-tu me rÃ©sumer ce quâ€™est lâ€™intelligence artificielle ?",
    "reponse": None
}
etat_final = app.invoke(etat_initial)
print("RÃ©ponse :", etat_final["reponse"])
```

```python
import PyPDF2

def extraire_texte_pdf(chemin_pdf):
    """
    Extrait tout le texte d'un fichier PDF
    """
    print(f"ğŸ”„ Lecture du fichier {chemin_pdf}...")

    with open(chemin_pdf, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        texte_complet = ""

        print(f"ğŸ“„ Nombre de pages : {len(reader.pages)}")

        for i, page in enumerate(reader.pages, 1):
            texte_complet += page.extract_text() + "\n"
            print(f"âœ… Page {i}/{len(reader.pages)} extraite")

    print(f"\nâœ… Extraction terminÃ©e ! Taille du texte : {len(texte_complet)} caractÃ¨res\n")
    return texte_complet

# Extraire le texte du PDF
texte_knowledge_base = extraire_texte_pdf("data_cleaning.pdf")

# Afficher un aperÃ§u
print("="*60)
print("APERÃ‡U DU CONTENU EXTRAIT :")
print("="*60)
print(texte_knowledge_base[:1000])  # Affiche les 1000 premiers caractÃ¨res
print("\n... (suite du texte) ...")
```
