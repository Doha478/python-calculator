````python
"""
Agentic ML Workflow - Architecture AmÃ©liorÃ©e
LangGraph + Mistral 7B + Tkinter

Installation pour Google Colab:
!pip install langgraph langchain langchain-community langchain-mistralai
!pip install scikit-learn pandas numpy matplotlib seaborn
"""

import os
import json
import pandas as pd
import numpy as np
from typing import TypedDict, Dict, Any
from dataclasses import dataclass

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, silhouette_score
import matplotlib.pyplot as plt

from langgraph.graph import StateGraph, END
from langchain_mistralai import ChatMistralAI


# =========================
# CONFIGURATION
# =========================

@dataclass
class WorkflowConfig:
    """Configuration centralisÃ©e du workflow"""
    model_name: str = "mistral-small-latest"
    temperature: float = 0
    default_k_neighbors: int = 3
    test_size: float = 0.2
    random_state: int = 42
    n_init_kmeans: int = 10
    max_k_clusters: int = 11
    min_k_clusters: int = 2


# =========================
# Ã‰TAT DU WORKFLOW
# =========================

class MLState(TypedDict):
    """Ã‰tat global du pipeline ML"""
    description: str
    dataset_path: str
    dataset: pd.DataFrame
    problem_type: str
    preprocessed_data: np.ndarray
    target: np.ndarray
    optimal_k: int
    results: dict
    error: str


# =========================
# UTILITAIRES
# =========================

class JSONParser:
    """Gestionnaire de parsing JSON depuis les rÃ©ponses LLM"""

    @staticmethod
    def parse(text: str) -> Dict[str, Any]:
        """Parse JSON en gÃ©rant les markdown fences"""
        try:
            cleaned_text = JSONParser._clean_markdown(text)
            json_content = JSONParser._extract_json_object(cleaned_text)
            return json.loads(json_content)
        except Exception as e:
            raise ValueError(f"Ã‰chec du parsing JSON. Sortie brute:\n{text}") from e

    @staticmethod
    def _clean_markdown(text: str) -> str:
        """Supprime les balises markdown"""
        if "```json" in text:
            return text.split("```json")[1].split("```")[0].strip()
        elif "```" in text:
            return text.split("```")[1].split("```")[0].strip()
        return text

    @staticmethod
    def _extract_json_object(text: str) -> str:
        """Extrait l'objet JSON du texte"""
        start = text.find("{")
        end = text.rfind("}") + 1

        if start == -1 or end == 0:
            raise ValueError("Aucun objet JSON trouvÃ©")

        return text[start:end]


class Logger:
    """Gestionnaire de logs avec emojis"""

    @staticmethod
    def info(message: str):
        print(f"â„¹ï¸  {message}")

    @staticmethod
    def success(message: str):
        print(f"âœ… {message}")

    @staticmethod
    def error(message: str):
        print(f"âŒ {message}")

    @staticmethod
    def warning(message: str):
        print(f"âš ï¸  {message}")

    @staticmethod
    def loading(message: str):
        print(f"ðŸ“‚ {message}")

    @staticmethod
    def robot(message: str):
        print(f"ðŸ¤– {message}")

    @staticmethod
    def rocket(message: str):
        print(f"ðŸš€ {message}")


# =========================
# PROCESSEURS DE DONNÃ‰ES
# =========================

class DataLoader:
    """Chargeur de datasets"""

    @staticmethod
    def load_csv(path: str) -> pd.DataFrame:
        """Charge un fichier CSV"""
        if not os.path.exists(path):
            raise FileNotFoundError(f"Fichier introuvable: {path}")
        return pd.read_csv(path)


class DataPreprocessor:
    """PrÃ©processeur de donnÃ©es"""

    def __init__(self):
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()

    def preprocess_features(self, df: pd.DataFrame) -> np.ndarray:
        """PrÃ©traite les features numÃ©riques"""
        numeric_df = df.select_dtypes(include=[np.number])
        numeric_df = numeric_df.fillna(numeric_df.mean())
        return self.scaler.fit_transform(numeric_df)

    def encode_target(self, target: np.ndarray) -> np.ndarray:
        """Encode la variable cible"""
        return self.label_encoder.fit_transform(target)


# =========================
# ANALYSEURS ML
# =========================

class ProblemAnalyzer:
    """Analyse le type de problÃ¨me ML avec Mistral"""

    def __init__(self, llm: ChatMistralAI):
        self.llm = llm

    def analyze(self, description: str, df: pd.DataFrame) -> Dict[str, Any]:
        """DÃ©termine le type de problÃ¨me"""
        prompt = self._build_prompt(description, df)
        response = self.llm.invoke(prompt)
        return JSONParser.parse(response.content.strip())

    def _build_prompt(self, description: str, df: pd.DataFrame) -> str:
        """Construit le prompt d'analyse"""
        return f"""
Vous Ãªtes un expert en Machine Learning.

DÃ©terminez s'il s'agit d'un problÃ¨me de classification ou de clustering.

Description utilisateur:
{description}

Colonnes du dataset:
{df.columns.tolist()}

Ã‰chantillon de donnÃ©es:
{df.head(3).to_string()}

RÃ©pondez UNIQUEMENT en format JSON:
{{
  "problem_type": "classification" ou "clustering",
  "target_column": "nom_colonne" ou null,
  "reasoning": "explication brÃ¨ve"
}}
"""


class HyperparameterOptimizer:
    """Optimiseur d'hyperparamÃ¨tres"""

    def __init__(self, config: WorkflowConfig):
        self.config = config

    def optimize_k_clustering(self, X: np.ndarray) -> int:
        """Trouve le K optimal pour clustering (mÃ©thode du coude)"""
        k_range = range(
            self.config.min_k_clusters,
            min(self.config.max_k_clusters, len(X))
        )
        scores = []

        for k in k_range:
            kmeans = KMeans(
                n_clusters=k,
                n_init=self.config.n_init_kmeans,
                random_state=self.config.random_state
            )
            labels = kmeans.fit_predict(X)
            scores.append(silhouette_score(X, labels))

        return k_range[np.argmax(scores)]


class MLExecutor:
    """ExÃ©cuteur d'algorithmes ML"""

    def __init__(self, config: WorkflowConfig):
        self.config = config

    def execute_clustering(self, X: np.ndarray, k: int) -> Dict[str, Any]:
        """ExÃ©cute KMeans clustering"""
        kmeans = KMeans(
            n_clusters=k,
            n_init=self.config.n_init_kmeans,
            random_state=self.config.random_state
        )
        labels = kmeans.fit_predict(X)

        return {
            "optimal_k": k,
            "silhouette_score": silhouette_score(X, labels),
            "labels": labels,
            "X": X
        }

    def execute_classification(self, X: np.ndarray, y: np.ndarray, k: int) -> Dict[str, Any]:
        """ExÃ©cute KNN classification"""
        X_train, X_test, y_train, y_test = train_test_split(
            X, y,
            test_size=self.config.test_size,
            random_state=self.config.random_state
        )

        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        predictions = knn.predict(X_test)

        return {
            "k": k,
            "accuracy": accuracy_score(y_test, predictions),
            "predictions": predictions,
            "actual": y_test,
            "X_test": X_test
        }


# =========================
# WORKFLOW PRINCIPAL
# =========================

class MLWorkflowEngine:
    """Moteur principal du workflow agentic"""

    def __init__(self, api_key: str, config: WorkflowConfig = None):
        self.config = config or WorkflowConfig()
        self.llm = ChatMistralAI(
            model=self.config.model_name,
            api_key=api_key,
            temperature=self.config.temperature
        )

        # Composants
        self.data_loader = DataLoader()
        self.preprocessor = DataPreprocessor()
        self.analyzer = ProblemAnalyzer(self.llm)
        self.optimizer = HyperparameterOptimizer(self.config)
        self.executor = MLExecutor(self.config)

        # Workflow LangGraph
        self.workflow = self._build_graph()

    def _build_graph(self) -> StateGraph:
        """Construction du graphe LangGraph"""
        graph = StateGraph(MLState)

        # Ajout des nÅ“uds
        graph.add_node("load", self._node_load_data)
        graph.add_node("analyze", self._node_analyze_problem)
        graph.add_node("preprocess", self._node_preprocess)
        graph.add_node("optimize", self._node_optimize)
        graph.add_node("execute", self._node_execute)

        # DÃ©finition du flux
        graph.set_entry_point("load")
        graph.add_edge("load", "analyze")
        graph.add_edge("analyze", "preprocess")
        graph.add_edge("preprocess", "optimize")
        graph.add_edge("optimize", "execute")
        graph.add_edge("execute", END)

        return graph.compile()

    # ========== NÅ’UDS DU GRAPHE ==========

    def _node_load_data(self, state: MLState) -> MLState:
        """NÅ“ud: Chargement des donnÃ©es"""
        try:
            Logger.loading("Chargement du dataset...")
            state["dataset"] = self.data_loader.load_csv(state["dataset_path"])
            state["error"] = ""
            Logger.success(
                f"Dataset chargÃ©: {len(state['dataset'])} lignes, "
                f"{len(state['dataset'].columns)} colonnes"
            )
        except Exception as e:
            state["error"] = f"Ã‰chec du chargement: {e}"
            Logger.error(state["error"])

        return state

    def _node_analyze_problem(self, state: MLState) -> MLState:
        """NÅ“ud: Analyse du problÃ¨me"""
        if state.get("error"):
            return state

        try:
            Logger.robot("Analyse du problÃ¨me avec Mistral 7B...")
            result = self.analyzer.analyze(state["description"], state["dataset"])

            state["problem_type"] = result["problem_type"].lower()
            Logger.success(f"Type de problÃ¨me dÃ©tectÃ©: {state['problem_type'].upper()}")
            Logger.info(f"Raisonnement: {result.get('reasoning', 'N/A')}")

            # Gestion de la colonne cible
            if state["problem_type"] == "classification":
                self._extract_target_column(state, result)

        except Exception as e:
            state["error"] = f"Ã‰chec de l'analyse: {e}"
            Logger.error(state["error"])

        return state

    def _node_preprocess(self, state: MLState) -> MLState:
        """NÅ“ud: PrÃ©traitement"""
        if state.get("error"):
            return state

        try:
            Logger.info("PrÃ©traitement des donnÃ©es...")
            state["preprocessed_data"] = self.preprocessor.preprocess_features(
                state["dataset"]
            )

            if state["problem_type"] == "classification" and state.get("target") is not None:
                state["target"] = self.preprocessor.encode_target(state["target"])

            Logger.success(f"Forme prÃ©traitÃ©e: {state['preprocessed_data'].shape}")

        except Exception as e:
            state["error"] = f"Ã‰chec du prÃ©traitement: {e}"
            Logger.error(state["error"])

        return state

    def _node_optimize(self, state: MLState) -> MLState:
        """NÅ“ud: Optimisation des hyperparamÃ¨tres"""
        if state.get("error"):
            return state

        try:
            Logger.info("Optimisation des hyperparamÃ¨tres...")

            if state["problem_type"] == "clustering":
                state["optimal_k"] = self.optimizer.optimize_k_clustering(
                    state["preprocessed_data"]
                )
                Logger.success(f"K optimal (clusters): {state['optimal_k']}")
            else:
                state["optimal_k"] = self.config.default_k_neighbors
                Logger.success(f"K utilisÃ© (voisins): {state['optimal_k']}")

        except Exception as e:
            state["error"] = f"Ã‰chec de l'optimisation: {e}"
            Logger.error(state["error"])

        return state

    def _node_execute(self, state: MLState) -> MLState:
        """NÅ“ud: ExÃ©cution de l'algorithme"""
        if state.get("error"):
            return state

        try:
            Logger.rocket("ExÃ©cution de l'algorithme...")

            if state["problem_type"] == "clustering":
                state["results"] = self.executor.execute_clustering(
                    state["preprocessed_data"],
                    state["optimal_k"]
                )
                Logger.success(
                    f"KMeans terminÃ© - Score Silhouette: "
                    f"{state['results']['silhouette_score']:.4f}"
                )
            else:
                state["results"] = self.executor.execute_classification(
                    state["preprocessed_data"],
                    state["target"],
                    state["optimal_k"]
                )
                Logger.success(
                    f"KNN terminÃ© - PrÃ©cision: "
                    f"{state['results']['accuracy']:.4f}"
                )

        except Exception as e:
            state["error"] = f"Ã‰chec de l'exÃ©cution: {e}"
            Logger.error(state["error"])

        return state

    # ========== HELPERS ==========

    def _extract_target_column(self, state: MLState, analysis: Dict[str, Any]):
        """Extrait la colonne cible pour classification"""
        target_col = analysis.get("target_column")
        df = state["dataset"]

        if target_col and target_col in df.columns:
            state["target"] = df[target_col].values
            state["dataset"] = df.drop(columns=[target_col])
            Logger.info(f"Colonne cible identifiÃ©e: {target_col}")
        else:
            Logger.warning(f"Colonne cible '{target_col}' introuvable, utilisation de la derniÃ¨re colonne")
            state["target"] = df.iloc[:, -1].values
            state["dataset"] = df.iloc[:, :-1]

    def run(self, description: str, dataset_path: str) -> MLState:
        """Lance le workflow complet"""
        initial_state = {
            "description": description,
            "dataset_path": dataset_path,
            "dataset": None,
            "problem_type": "",
            "preprocessed_data": None,
            "target": None,
            "optimal_k": 0,
            "results": {},
            "error": ""
        }

        return self.workflow.invoke(initial_state)


# =========================
# VISUALISATION
# =========================

class ResultVisualizer:
    """Visualiseur de rÃ©sultats ML"""

    @staticmethod
    def visualize(state: MLState):
        """CrÃ©e les visualisations appropriÃ©es"""
        if state.get("error"):
            return

        if state["problem_type"] == "clustering":
            ResultVisualizer._plot_clustering(state["results"])
        else:
            ResultVisualizer._plot_classification(state["results"])

    @staticmethod
    def _plot_clustering(results: Dict[str, Any]):
        """Visualisation clustering"""
        X = results["X"]
        labels = results["labels"]
        k = results["optimal_k"]

        X_2d = ResultVisualizer._reduce_dimensions(X)
        xlabel, ylabel = ResultVisualizer._get_axis_labels(X)

        plt.figure(figsize=(10, 6))
        scatter = plt.scatter(
            X_2d[:, 0], X_2d[:, 1],
            c=labels, cmap='viridis',
            alpha=0.6, s=50
        )
        plt.colorbar(scatter, label='Cluster')
        plt.title(
            f'RÃ©sultats KMeans Clustering (k={k})\n'
            f'Score Silhouette: {results["silhouette_score"]:.4f}'
        )
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    @staticmethod
    def _plot_classification(results: Dict[str, Any]):
        """Visualisation classification"""
        X = results["X_test"]
        preds = results["predictions"]
        k = results["k"]

        X_2d = ResultVisualizer._reduce_dimensions(X)
        xlabel, ylabel = ResultVisualizer._get_axis_labels(X)

        plt.figure(figsize=(10, 6))
        scatter = plt.scatter(
            X_2d[:, 0], X_2d[:, 1],
            c=preds, cmap='coolwarm',
            alpha=0.6, s=50
        )
        plt.colorbar(scatter, label='Classe PrÃ©dite')
        plt.title(
            f'RÃ©sultats KNN Classification (k={k})\n'
            f'PrÃ©cision: {results["accuracy"]:.4f}'
        )
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    @staticmethod
    def _reduce_dimensions(X: np.ndarray) -> np.ndarray:
        """RÃ©duit Ã  2D si nÃ©cessaire"""
        if X.shape[1] > 2:
            pca = PCA(n_components=2)
            return pca.fit_transform(X)
        return X[:, :2]

    @staticmethod
    def _get_axis_labels(X: np.ndarray) -> tuple:
        """Retourne les labels d'axes appropriÃ©s"""
        if X.shape[1] > 2:
            return "Composante Principale 1", "Composante Principale 2"
        return "Feature 1", "Feature 2"


class ResultPrinter:
    """Afficheur de rÃ©sultats dÃ©taillÃ©s"""

    @staticmethod
    def print_results(state: MLState):
        """Affiche les rÃ©sultats complets"""
        if state.get("error"):
            ResultPrinter._print_error(state["error"])
            return

        ResultPrinter._print_header()
        ResultPrinter._print_problem_type(state["problem_type"])

        if state["problem_type"] == "clustering":
            ResultPrinter._print_clustering_results(state["results"])
        else:
            ResultPrinter._print_classification_results(state["results"])

        ResultPrinter._print_footer()

    @staticmethod
    def _print_error(error: str):
        print("\n" + "="*60)
        print("âŒ Ã‰CHEC DU WORKFLOW")
        print("="*60)
        print(f"Erreur: {error}")

    @staticmethod
    def _print_header():
        print("\n" + "="*60)
        print("ðŸŽ‰ WORKFLOW TERMINÃ‰ AVEC SUCCÃˆS")
        print("="*60)

    @staticmethod
    def _print_problem_type(problem_type: str):
        print(f"\nðŸ“Š Type de ProblÃ¨me: {problem_type.upper()}")
        print("-"*60)

    @staticmethod
    def _print_clustering_results(results: Dict[str, Any]):
        print(f"âœ… K optimal (clusters): {results['optimal_k']}")
        print(f"ðŸ“ˆ Score Silhouette: {results['silhouette_score']:.4f}")

        unique, counts = np.unique(results['labels'], return_counts=True)
        print(f"\nðŸ“¦ Distribution des Clusters:")
        for cluster, count in zip(unique, counts):
            percentage = count / len(results['labels']) * 100
            print(f"   Cluster {cluster}: {count} Ã©chantillons ({percentage:.1f}%)")

    @staticmethod
    def _print_classification_results(results: Dict[str, Any]):
        print(f"âœ… K (voisins): {results['k']}")
        print(f"ðŸŽ¯ PrÃ©cision: {results['accuracy']:.4f}")

        unique, counts = np.unique(results['predictions'], return_counts=True)
        print(f"\nðŸ“¦ Distribution des Classes PrÃ©dites:")
        for cls, count in zip(unique, counts):
            percentage = count / len(results['predictions']) * 100
            print(f"   Classe {cls}: {count} Ã©chantillons ({percentage:.1f}%)")

        most_common = unique[np.argmax(counts)]
        print(f"\nðŸ† Classe la Plus FrÃ©quente: {most_common}")

    @staticmethod
    def _print_footer():
        print("\n" + "="*60)


# =========================
# INTERFACE CLI
# =========================

class CLIInterface:
    """Interface en ligne de commande"""

    @staticmethod
    def run():
        """Lance le pipeline en mode CLI"""
        CLIInterface._print_banner()

        # Collecte des inputs
        api_key = CLIInterface._get_api_key()
        if not api_key:
            return

        dataset_path = CLIInterface._get_dataset_path()
        if not dataset_path:
            return

        description = CLIInterface._get_description()
        if not description:
            return

        CLIInterface._print_start_banner()

        # ExÃ©cution du workflow
        try:
            engine = MLWorkflowEngine(api_key)
            state = engine.run(description, dataset_path)

            # Affichage des rÃ©sultats
            ResultPrinter.print_results(state)

            # Visualisation
            if not state.get("error"):
                print("\nðŸ“Š GÃ©nÃ©ration de la visualisation...")
                ResultVisualizer.visualize(state)

        except Exception as e:
            Logger.error(f"ERREUR FATALE: {e}")
            import traceback
            traceback.print_exc()

    @staticmethod
    def _print_banner():
        print("="*60)
        print("ðŸ¤– WORKFLOW ML AGENTIC â€“ MODE CLI")
        print("   PropulsÃ© par Mistral 7B + LangGraph")
        print("="*60)

    @staticmethod
    def _get_api_key() -> str:
        api_key = input("\nðŸ”‘ Entrez votre clÃ© API Mistral: ").strip()
        if not api_key:
            Logger.error("La clÃ© API est requise")
        return api_key

    @staticmethod
    def _get_dataset_path() -> str:
        path = input("ðŸ“‚ Entrez le chemin du fichier CSV: ").strip()
        if not os.path.exists(path):
            Logger.error(f"Fichier introuvable: {path}")
            return ""
        return path

    @staticmethod
    def _get_description() -> str:
        print("\nðŸ“ DÃ©crivez votre problÃ¨me ML:")
        description = input("> ").strip()
        if not description:
            Logger.error("La description est requise")
        return description

    @staticmethod
    def _print_start_banner():
        print("\n" + "="*60)
        print("ðŸš€ DÃ‰MARRAGE DU WORKFLOW...")
        print("="*60 + "\n")


# =========================
# POINT D'ENTRÃ‰E
# =========================

if __name__ == "__main__":
    CLIInterface.run()
````

````python
"""
Agentic ML Workflow v2 - Nouveaux Datasets
LangGraph + Mistral 7B
Datasets: SantÃ©, Finance, Retail

Installation pour Google Colab:
!pip install langgraph langchain langchain-community langchain-mistralai
!pip install scikit-learn pandas numpy matplotlib seaborn
!pip install python-dotenv
"""

import os
import json
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, silhouette_score
import matplotlib.pyplot as plt
from dotenv import load_dotenv

from typing import TypedDict
from langgraph.graph import StateGraph, END
from langchain_mistralai import ChatMistralAI


# =========================
# CRÃ‰ATION DES NOUVEAUX DATASETS
# =========================

def create_all_datasets():
    """CrÃ©e tous les nouveaux datasets"""
    print("ðŸš€ CrÃ©ation des nouveaux datasets...")
    print("="*60)

    # 1. DATASET SANTÃ‰ - Clustering de profils de patients
    create_health_clustering()

    # 2. DATASET FINANCE - Classification de crÃ©dits
    create_credit_classification()

    # 3. DATASET RETAIL - Clustering de comportements clients
    create_customer_clustering()

    # 4. DATASET IMMOBILIER - Classification de prix
    create_real_estate_classification()

    print("\n" + "="*60)
    print("ðŸŽ‰ TOUS LES DATASETS ONT Ã‰TÃ‰ CRÃ‰Ã‰S !")
    print("="*60)


def create_health_clustering():
    """Dataset 1: Clustering de profils de santÃ©"""
    print("\nðŸ“Š CrÃ©ation de health_profiles.csv...")

    np.random.seed(42)
    data = []

    # Groupe 1: Patients jeunes en bonne santÃ©
    for _ in range(60):
        data.append({
            'age': np.random.randint(20, 35),
            'blood_pressure': np.random.randint(110, 130),
            'cholesterol': np.random.randint(150, 200),
            'bmi': round(np.random.uniform(18.5, 24.9), 1),
            'exercise_hours_week': np.random.randint(3, 8)
        })

    # Groupe 2: Patients d'Ã¢ge moyen Ã  risque modÃ©rÃ©
    for _ in range(60):
        data.append({
            'age': np.random.randint(40, 55),
            'blood_pressure': np.random.randint(130, 150),
            'cholesterol': np.random.randint(200, 250),
            'bmi': round(np.random.uniform(25, 29.9), 1),
            'exercise_hours_week': np.random.randint(1, 4)
        })

    # Groupe 3: Patients Ã¢gÃ©s Ã  haut risque
    for _ in range(60):
        data.append({
            'age': np.random.randint(60, 80),
            'blood_pressure': np.random.randint(140, 180),
            'cholesterol': np.random.randint(240, 300),
            'bmi': round(np.random.uniform(30, 40), 1),
            'exercise_hours_week': np.random.randint(0, 2)
        })

    df = pd.DataFrame(data)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    df.to_csv('health_profiles.csv', index=False)

    print(f"âœ… health_profiles.csv crÃ©Ã© ({len(df)} lignes)")
    print(f"   Colonnes: {list(df.columns)}")
    print(f"   Description: 'Groupe les profils de santÃ© similaires'")


def create_credit_classification():
    """Dataset 2: Classification de risque de crÃ©dit"""
    print("\nðŸ“Š CrÃ©ation de credit_risk.csv...")

    np.random.seed(42)
    data = []

    # Faible risque
    for _ in range(70):
        data.append({
            'income': np.random.randint(50000, 120000),
            'credit_score': np.random.randint(700, 850),
            'debt_ratio': round(np.random.uniform(0.1, 0.3), 2),
            'employment_years': np.random.randint(5, 20),
            'loan_amount': np.random.randint(5000, 30000),
            'risk_level': 'Low'
        })

    # Risque moyen
    for _ in range(70):
        data.append({
            'income': np.random.randint(30000, 60000),
            'credit_score': np.random.randint(600, 720),
            'debt_ratio': round(np.random.uniform(0.3, 0.5), 2),
            'employment_years': np.random.randint(2, 8),
            'loan_amount': np.random.randint(10000, 50000),
            'risk_level': 'Medium'
        })

    # Haut risque
    for _ in range(60):
        data.append({
            'income': np.random.randint(20000, 40000),
            'credit_score': np.random.randint(500, 620),
            'debt_ratio': round(np.random.uniform(0.5, 0.8), 2),
            'employment_years': np.random.randint(0, 3),
            'loan_amount': np.random.randint(20000, 80000),
            'risk_level': 'High'
        })

    df = pd.DataFrame(data)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    df.to_csv('credit_risk.csv', index=False)

    print(f"âœ… credit_risk.csv crÃ©Ã© ({len(df)} lignes)")
    print(f"   Colonnes: {list(df.columns)}")
    print(f"   Classes: {df['risk_level'].unique().tolist()}")
    print(f"   Description: 'PrÃ©dis le niveau de risque de crÃ©dit'")


def create_customer_clustering():
    """Dataset 3: Clustering de comportements clients retail"""
    print("\nðŸ“Š CrÃ©ation de customer_behavior.csv...")

    np.random.seed(42)
    data = []

    # Segment 1: Acheteurs occasionnels
    for _ in range(55):
        data.append({
            'monthly_purchases': np.random.randint(1, 3),
            'avg_transaction': np.random.randint(20, 60),
            'website_visits': np.random.randint(2, 8),
            'loyalty_points': np.random.randint(0, 100),
            'customer_tenure_months': np.random.randint(1, 12)
        })

    # Segment 2: Clients rÃ©guliers
    for _ in range(55):
        data.append({
            'monthly_purchases': np.random.randint(4, 10),
            'avg_transaction': np.random.randint(50, 120),
            'website_visits': np.random.randint(10, 25),
            'loyalty_points': np.random.randint(100, 500),
            'customer_tenure_months': np.random.randint(12, 36)
        })

    # Segment 3: VIP / Gros acheteurs
    for _ in range(50):
        data.append({
            'monthly_purchases': np.random.randint(10, 25),
            'avg_transaction': np.random.randint(120, 300),
            'website_visits': np.random.randint(25, 50),
            'loyalty_points': np.random.randint(500, 2000),
            'customer_tenure_months': np.random.randint(36, 120)
        })

    df = pd.DataFrame(data)
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    df.to_csv('customer_behavior.csv', index=False)

    print(f"âœ… customer_behavior.csv crÃ©Ã© ({len(df)} lignes)")
    print(f"   Colonnes: {list(df.columns)}")
    print(f"   Description: 'Segmente les clients par comportement d achat'")


def create_real_estate_classification():
    """Dataset 4: Classification de gamme de prix immobilier"""
    print("\nðŸ“Š CrÃ©ation de real_estate.csv...")

    np.random.seed(42)
    data = []

    # Ã‰conomique
    for _ in range(65):
        data.append({
            'surface_m2': np.random.randint(30, 70),
            'num_rooms': np.random.randint(1, 3),
            'distance_center_km': round(np.random.uniform(8, 20), 1),
            'year_built': np.random.randint(1960, 1990),
            'energy_rating': np.random.choice(['D', 'E', 'F', 'G']),
            'price_range': 'Affordable'
        })

    # Moyen de gamme
    for _ in range(70):
        data.append({
            'surface_m2': np.random.randint(60, 110),
            'num_rooms': np.random.randint(2, 4),
            'distance_center_km': round(np.random.uniform(4, 12), 1),
            'year_built': np.random.randint(1985, 2010),
            'energy_rating': np.random.choice(['B', 'C', 'D']),
            'price_range': 'Mid-Range'
        })

    # Luxe
    for _ in range(65):
        data.append({
            'surface_m2': np.random.randint(100, 250),
            'num_rooms': np.random.randint(4, 8),
            'distance_center_km': round(np.random.uniform(0.5, 6), 1),
            'year_built': np.random.randint(2005, 2024),
            'energy_rating': np.random.choice(['A', 'A+', 'B']),
            'price_range': 'Luxury'
        })

    df = pd.DataFrame(data)
    # Convertir energy_rating en numÃ©rique pour le ML
    rating_map = {'G': 1, 'F': 2, 'E': 3, 'D': 4, 'C': 5, 'B': 6, 'A': 7, 'A+': 8}
    df['energy_score'] = df['energy_rating'].map(rating_map)
    df = df.drop('energy_rating', axis=1)

    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    df.to_csv('real_estate.csv', index=False)

    print(f"âœ… real_estate.csv crÃ©Ã© ({len(df)} lignes)")
    print(f"   Colonnes: {list(df.columns)}")
    print(f"   Classes: {df['price_range'].unique().tolist()}")
    print(f"   Description: 'Classifie les biens immobiliers par gamme de prix'")


# =========================
# HELPER FUNCTIONS
# =========================

def safe_json_parse(text: str):
    """Parse JSON from LLM response, handling markdown fences"""
    try:
        if "```json" in text:
            text = text.split("```json")[1].split("```")[0].strip()
        elif "```" in text:
            text = text.split("```")[1].split("```")[0].strip()

        start = text.find("{")
        end = text.rfind("}") + 1

        if start == -1 or end == 0:
            raise ValueError("No JSON object found")

        return json.loads(text[start:end])

    except Exception as e:
        raise ValueError(f"LLM JSON parsing failed. Raw output:\n{text}") from e


# =========================
# STATE DEFINITION
# =========================

class MLState(TypedDict):
    description: str
    dataset_path: str
    dataset: pd.DataFrame
    problem_type: str
    preprocessed_data: np.ndarray
    target: np.ndarray
    optimal_k: int
    results: dict
    error: str


# =========================
# AGENTIC WORKFLOW
# =========================

class MLAgenticWorkflow:
    def __init__(self, mistral_api_key: str):
        self.llm = ChatMistralAI(
            model="mistral-small-latest",
            api_key=mistral_api_key,
            temperature=0
        )
        self.workflow = self._build_workflow()

    def _build_workflow(self):
        """Build LangGraph workflow"""
        workflow = StateGraph(MLState)

        workflow.add_node("load_data", self.load_data_node)
        workflow.add_node("analyze_problem", self.analyze_problem_node)
        workflow.add_node("preprocess_data", self.preprocess_data_node)
        workflow.add_node("optimize_hyperparameters", self.optimize_hyperparameters_node)
        workflow.add_node("execute_algorithm", self.execute_algorithm_node)

        workflow.set_entry_point("load_data")
        workflow.add_edge("load_data", "analyze_problem")
        workflow.add_edge("analyze_problem", "preprocess_data")
        workflow.add_edge("preprocess_data", "optimize_hyperparameters")
        workflow.add_edge("optimize_hyperparameters", "execute_algorithm")
        workflow.add_edge("execute_algorithm", END)

        return workflow.compile()

    def load_data_node(self, state: MLState) -> MLState:
        """Node 1: Load CSV dataset"""
        try:
            print("ðŸ“‚ Chargement du dataset...")
            state["dataset"] = pd.read_csv(state["dataset_path"])
            state["error"] = ""
            print(f"âœ… ChargÃ©: {len(state['dataset'])} lignes, {len(state['dataset'].columns)} colonnes")
        except Exception as e:
            state["error"] = f"Ã‰chec du chargement: {e}"
            print(f"âŒ {state['error']}")
        return state

    def analyze_problem_node(self, state: MLState) -> MLState:
        """Node 2: Use Mistral LLM to analyze problem type"""
        if state.get("error"):
            return state

        try:
            print("ðŸ¤– Analyse du problÃ¨me avec Mistral 7B...")
            df = state["dataset"]

            prompt = f"""
Tu es un expert en Machine Learning.

DÃ©termine s'il s'agit d'un problÃ¨me de classification ou de clustering.

Description utilisateur:
{state["description"]}

Colonnes du dataset:
{df.columns.tolist()}

Ã‰chantillon de donnÃ©es:
{df.head(3).to_string()}

RÃ©ponds UNIQUEMENT en format JSON:
{{
  "problem_type": "classification" ou "clustering",
  "target_column": "nom_colonne" ou null,
  "reasoning": "explication brÃ¨ve"
}}
"""
            response = self.llm.invoke(prompt)
            content = response.content.strip()

            result = safe_json_parse(content)

            state["problem_type"] = result["problem_type"].lower()
            print(f"âœ… Type de problÃ¨me dÃ©tectÃ©: {state['problem_type'].upper()}")
            print(f"ðŸ’¡ Raisonnement: {result.get('reasoning', 'N/A')}")

            if state["problem_type"] == "classification" and result.get("target_column"):
                target_col = result["target_column"]
                if target_col in df.columns:
                    state["target"] = df[target_col].values
                    state["dataset"] = df.drop(columns=[target_col])
                    print(f"ðŸŽ¯ Colonne cible identifiÃ©e: {target_col}")
                else:
                    print(f"âš ï¸ Colonne cible '{target_col}' introuvable, utilisation de la derniÃ¨re colonne")
                    state["target"] = df.iloc[:, -1].values
                    state["dataset"] = df.iloc[:, :-1]

        except Exception as e:
            state["error"] = f"Ã‰chec de l'analyse: {e}"
            print(f"âŒ {state['error']}")

        return state

    def preprocess_data_node(self, state: MLState) -> MLState:
        """Node 3: Handle missing values and scaling"""
        if state.get("error"):
            return state

        try:
            print("ðŸ”§ PrÃ©traitement des donnÃ©es...")
            df = state["dataset"].select_dtypes(include=[np.number])

            df = df.fillna(df.mean())

            scaler = StandardScaler()
            state["preprocessed_data"] = scaler.fit_transform(df)

            if state["problem_type"] == "classification" and state.get("target") is not None:
                le = LabelEncoder()
                state["target"] = le.fit_transform(state["target"])

            print(f"âœ… Forme prÃ©traitÃ©e: {state['preprocessed_data'].shape}")

        except Exception as e:
            state["error"] = f"Ã‰chec du prÃ©traitement: {e}"
            print(f"âŒ {state['error']}")

        return state

    def optimize_hyperparameters_node(self, state: MLState) -> MLState:
        """Node 4: Optimize k using elbow method for clustering"""
        if state.get("error"):
            return state

        try:
            print("âš™ï¸ Optimisation des hyperparamÃ¨tres...")

            if state["problem_type"] == "clustering":
                X = state["preprocessed_data"]
                scores = []
                k_range = range(2, min(11, len(X)))

                for k in k_range:
                    km = KMeans(n_clusters=k, n_init=10, random_state=42)
                    labels = km.fit_predict(X)
                    scores.append(silhouette_score(X, labels))

                state["optimal_k"] = k_range[np.argmax(scores)]
                print(f"âœ… K optimal (clusters): {state['optimal_k']}")
            else:
                state["optimal_k"] = 3
                print(f"âœ… K utilisÃ© (voisins): {state['optimal_k']}")

        except Exception as e:
            state["error"] = f"Ã‰chec de l'optimisation: {e}"
            print(f"âŒ {state['error']}")

        return state

    def execute_algorithm_node(self, state: MLState) -> MLState:
        """Node 5: Execute KMeans or KNN"""
        if state.get("error"):
            return state

        try:
            print("ðŸš€ ExÃ©cution de l'algorithme...")
            X = state["preprocessed_data"]
            k = state["optimal_k"]

            if state["problem_type"] == "clustering":
                km = KMeans(n_clusters=k, n_init=10, random_state=42)
                labels = km.fit_predict(X)

                state["results"] = {
                    "optimal_k": k,
                    "silhouette_score": silhouette_score(X, labels),
                    "labels": labels,
                    "X": X
                }
                print(f"âœ… KMeans terminÃ© - Score Silhouette: {state['results']['silhouette_score']:.4f}")

            else:
                Xtr, Xte, ytr, yte = train_test_split(
                    X, state["target"], test_size=0.2, random_state=42
                )
                knn = KNeighborsClassifier(n_neighbors=k)
                knn.fit(Xtr, ytr)
                preds = knn.predict(Xte)

                state["results"] = {
                    "k": k,
                    "accuracy": accuracy_score(yte, preds),
                    "predictions": preds,
                    "actual": yte,
                    "X_test": Xte
                }
                print(f"âœ… KNN terminÃ© - PrÃ©cision: {state['results']['accuracy']:.4f}")

        except Exception as e:
            state["error"] = f"Ã‰chec de l'exÃ©cution: {e}"
            print(f"âŒ {state['error']}")

        return state

    def run(self, description: str, dataset_path: str) -> MLState:
        """Run the complete workflow"""
        return self.workflow.invoke({
            "description": description,
            "dataset_path": dataset_path,
            "dataset": None,
            "problem_type": "",
            "preprocessed_data": None,
            "target": None,
            "optimal_k": 0,
            "results": {},
            "error": ""
        })


# =========================
# VISUALIZATION
# =========================

def visualize_results(state: MLState):
    """Create visualizations for results"""
    if state.get("error"):
        return

    results = state["results"]
    problem_type = state["problem_type"]

    if problem_type == "clustering":
        X = results["X"]
        labels = results["labels"]
        k = results["optimal_k"]

        if X.shape[1] > 2:
            pca = PCA(n_components=2)
            X_2d = pca.fit_transform(X)
            xlabel, ylabel = "Composante Principale 1", "Composante Principale 2"
        else:
            X_2d = X[:, :2]
            xlabel, ylabel = "Feature 1", "Feature 2"

        plt.figure(figsize=(10, 6))
        scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap='viridis', alpha=0.6, s=50)
        plt.colorbar(scatter, label='Cluster')
        plt.title(f'RÃ©sultats KMeans Clustering (k={k})\nScore Silhouette: {results["silhouette_score"]:.4f}')
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    else:
        X = results["X_test"]
        preds = results["predictions"]
        k = results["k"]

        if X.shape[1] > 2:
            pca = PCA(n_components=2)
            X_2d = pca.fit_transform(X)
            xlabel, ylabel = "Composante Principale 1", "Composante Principale 2"
        else:
            X_2d = X[:, :2]
            xlabel, ylabel = "Feature 1", "Feature 2"

        plt.figure(figsize=(10, 6))
        scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=preds, cmap='coolwarm', alpha=0.6, s=50)
        plt.colorbar(scatter, label='Classe PrÃ©dite')
        plt.title(f'RÃ©sultats KNN Classification (k={k})\nPrÃ©cision: {results["accuracy"]:.4f}')
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()


def print_results(state: MLState):
    """Print detailed results"""
    if state.get("error"):
        print("\n" + "="*60)
        print("âŒ Ã‰CHEC DU WORKFLOW")
        print("="*60)
        print(f"Erreur: {state['error']}")
        return

    print("\n" + "="*60)
    print("ðŸŽ‰ WORKFLOW TERMINÃ‰ AVEC SUCCÃˆS")
    print("="*60)

    print(f"\nðŸ“Š Type de ProblÃ¨me: {state['problem_type'].upper()}")
    print("-"*60)

    results = state["results"]

    if state["problem_type"] == "clustering":
        print(f"âœ… K optimal (clusters): {results['optimal_k']}")
        print(f"ðŸ“ˆ Score Silhouette: {results['silhouette_score']:.4f}")

        unique, counts = np.unique(results['labels'], return_counts=True)
        print(f"\nðŸ“¦ Distribution des Clusters:")
        for cluster, count in zip(unique, counts):
            print(f"   Cluster {cluster}: {count} Ã©chantillons ({count/len(results['labels'])*100:.1f}%)")

    else:
        print(f"âœ… K (voisins): {results['k']}")
        print(f"ðŸŽ¯ PrÃ©cision: {results['accuracy']:.4f}")

        unique, counts = np.unique(results['predictions'], return_counts=True)
        print(f"\nðŸ“¦ Distribution des Classes PrÃ©dites:")
        for cls, count in zip(unique, counts):
            print(f"   Classe {cls}: {count} Ã©chantillons ({count/len(results['predictions'])*100:.1f}%)")

        most_common = unique[np.argmax(counts)]
        print(f"\nðŸ† Classe la Plus FrÃ©quente: {most_common}")

    print("\n" + "="*60)


# =========================
# CLI RUNNER
# =========================

def run_ml_pipeline_cli():
    """Run the ML pipeline in CLI mode"""
    print("="*60)
    print("ðŸ¤– WORKFLOW ML AGENTIC v2 â€“ MODE CLI")
    print("   PropulsÃ© par Mistral 7B + LangGraph")
    print("="*60)

    load_dotenv('myenv')
    api_key = os.getenv('MISTRAL_API_KEY')

    if api_key:
        print("\nâœ… ClÃ© API chargÃ©e depuis myenv")
    else:
        print("\nâš ï¸ MISTRAL_API_KEY introuvable dans myenv")
        api_key = input("ðŸ”‘ Entre ta clÃ© API Mistral: ").strip()
        if not api_key:
            print("âŒ La clÃ© API est requise")
            return

    dataset_path = input("ðŸ“‚ Entre le chemin du fichier CSV: ").strip()
    if not os.path.exists(dataset_path):
        print(f"âŒ Fichier introuvable: {dataset_path}")
        return

    print("\nðŸ“ DÃ©cris ton problÃ¨me ML:")
    description = input("> ").strip()
    if not description:
        print("âŒ La description est requise")
        return

    print("\n" + "="*60)
    print("ðŸš€ DÃ‰MARRAGE DU WORKFLOW...")
    print("="*60 + "\n")

    try:
        workflow = MLAgenticWorkflow(api_key)
        state = workflow.run(description, dataset_path)

        print_results(state)

        if not state.get("error"):
            print("\nðŸ“Š GÃ©nÃ©ration de la visualisation...")
            visualize_results(state)

    except Exception as e:
        print(f"\nâŒ ERREUR FATALE: {e}")
        import traceback
        traceback.print_exc()


# =========================
# ENTRY POINT
# =========================

if __name__ == "__main__":
    # D'abord crÃ©er les datasets
    print("ðŸŽ¬ Ã‰TAPE 1: CrÃ©ation des datasets")
    print("Veux-tu crÃ©er les nouveaux datasets? (o/n)")
    choice = input("> ").strip().lower()

    if choice == 'o' or choice == 'oui' or choice == 'y' or choice == 'yes':
        create_all_datasets()
        print("\nâœ… Datasets crÃ©Ã©s ! Tu peux maintenant lancer le workflow.\n")

    print("\nðŸŽ¬ Ã‰TAPE 2: Lancement du workflow ML")
    run_ml_pipeline_cli()
````

````python
"""
Agentic ML Workflow with LangGraph, Mistral 7B
Enhanced Visual Dashboard - Google Colab Ready

Installation Instructions for Google Colab:
!pip install langgraph langchain langchain-community langchain-mistralai
!pip install scikit-learn pandas numpy matplotlib seaborn
!pip install python-dotenv
"""

import os
import json
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, silhouette_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv

from typing import TypedDict
from langgraph.graph import StateGraph, END
from langchain_mistralai import ChatMistralAI

# Set style for beautiful plots
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")


# =========================
# HELPER FUNCTIONS
# =========================

def safe_json_parse(text: str):
    """Parse JSON from LLM response, handling markdown fences"""
    try:
        # Remove markdown fences if present
        if "```json" in text:
            text = text.split("```json")[1].split("```")[0].strip()
        elif "```" in text:
            text = text.split("```")[1].split("```")[0].strip()

        # Extract JSON object
        start = text.find("{")
        end = text.rfind("}") + 1

        if start == -1 or end == 0:
            raise ValueError("No JSON object found")

        return json.loads(text[start:end])

    except Exception as e:
        raise ValueError(f"LLM JSON parsing failed. Raw output:\n{text}") from e


# =========================
# STATE DEFINITION
# =========================

class MLState(TypedDict):
    description: str
    dataset_path: str
    dataset: pd.DataFrame
    original_dataset: pd.DataFrame
    problem_type: str
    llm_reasoning: str
    target_column: str
    preprocessed_data: np.ndarray
    target: np.ndarray
    optimal_k: int
    k_scores: list
    results: dict
    error: str


# =========================
# AGENTIC WORKFLOW
# =========================

class MLAgenticWorkflow:
    def __init__(self, mistral_api_key: str):
        self.llm = ChatMistralAI(
            model="mistral-small-latest",
            api_key=mistral_api_key,
            temperature=0
        )
        self.workflow = self._build_workflow()

    def _build_workflow(self):
        """Build LangGraph workflow"""
        workflow = StateGraph(MLState)

        workflow.add_node("load_data", self.load_data_node)
        workflow.add_node("analyze_problem", self.analyze_problem_node)
        workflow.add_node("preprocess_data", self.preprocess_data_node)
        workflow.add_node("optimize_hyperparameters", self.optimize_hyperparameters_node)
        workflow.add_node("execute_algorithm", self.execute_algorithm_node)

        workflow.set_entry_point("load_data")
        workflow.add_edge("load_data", "analyze_problem")
        workflow.add_edge("analyze_problem", "preprocess_data")
        workflow.add_edge("preprocess_data", "optimize_hyperparameters")
        workflow.add_edge("optimize_hyperparameters", "execute_algorithm")
        workflow.add_edge("execute_algorithm", END)

        return workflow.compile()

    def load_data_node(self, state: MLState) -> MLState:
        """Node 1: Load CSV dataset"""
        try:
            print("ðŸ“‚ Loading dataset...")
            df = pd.read_csv(state["dataset_path"])
            state["dataset"] = df.copy()
            state["original_dataset"] = df.copy()
            state["error"] = ""
            print(f"âœ… Loaded {len(df)} rows, {len(df.columns)} columns")
        except Exception as e:
            state["error"] = f"Data loading failed: {e}"
            print(f"âŒ {state['error']}")
        return state

    def analyze_problem_node(self, state: MLState) -> MLState:
        """Node 2: Use Mistral LLM to analyze problem type"""
        if state.get("error"):
            return state

        try:
            print("ðŸ¤– Analyzing problem with Mistral 7B...")
            df = state["dataset"]

            prompt = f"""
You are an ML expert.

Determine if this is a classification or clustering problem.

User description:
{state["description"]}

Dataset columns:
{df.columns.tolist()}

Sample data:
{df.head(3).to_string()}

Dataset info:
- Rows: {len(df)}
- Columns: {len(df.columns)}
- Numeric columns: {len(df.select_dtypes(include=[np.number]).columns)}

Respond ONLY in JSON format:
{{
  "problem_type": "classification" or "clustering",
  "target_column": "column_name" or null,
  "reasoning": "detailed explanation of why this is classification/clustering and what patterns you observed"
}}
"""

            response = self.llm.invoke(prompt)
            content = response.content.strip()

            result = safe_json_parse(content)

            state["problem_type"] = result["problem_type"].lower()
            state["llm_reasoning"] = result.get("reasoning", "No reasoning provided")
            state["target_column"] = result.get("target_column", "")

            print(f"âœ… Detected problem type: {state['problem_type'].upper()}")
            print(f"ðŸ’¡ LLM Reasoning: {state['llm_reasoning'][:100]}...")

            if state["problem_type"] == "classification" and result.get("target_column"):
                target_col = result["target_column"]
                if target_col in df.columns:
                    state["target"] = df[target_col].values
                    state["dataset"] = df.drop(columns=[target_col])
                    print(f"ðŸŽ¯ Target column identified: {target_col}")
                else:
                    print(f"âš ï¸ Target column '{target_col}' not found, using last column")
                    state["target"] = df.iloc[:, -1].values
                    state["dataset"] = df.iloc[:, :-1]
                    state["target_column"] = df.columns[-1]

        except Exception as e:
            state["error"] = f"Problem analysis failed: {e}"
            print(f"âŒ {state['error']}")

        return state

    def preprocess_data_node(self, state: MLState) -> MLState:
        """Node 3: Handle missing values and scaling"""
        if state.get("error"):
            return state

        try:
            print("ðŸ”§ Preprocessing data...")
            df = state["dataset"].select_dtypes(include=[np.number])

            # Handle missing values
            df = df.fillna(df.mean())

            # Scale features
            scaler = StandardScaler()
            state["preprocessed_data"] = scaler.fit_transform(df)

            # Encode target for classification
            if state["problem_type"] == "classification" and state.get("target") is not None:
                le = LabelEncoder()
                state["target"] = le.fit_transform(state["target"])

            print(f"âœ… Preprocessed shape: {state['preprocessed_data'].shape}")

        except Exception as e:
            state["error"] = f"Preprocessing failed: {e}"
            print(f"âŒ {state['error']}")

        return state

    def optimize_hyperparameters_node(self, state: MLState) -> MLState:
        """Node 4: Optimize k using elbow method for clustering"""
        if state.get("error"):
            return state

        try:
            print("âš™ï¸ Optimizing hyperparameters...")

            if state["problem_type"] == "clustering":
                X = state["preprocessed_data"]
                scores = []
                inertias = []
                k_range = range(2, min(11, len(X)))

                for k in k_range:
                    km = KMeans(n_clusters=k, n_init=10, random_state=42)
                    labels = km.fit_predict(X)
                    scores.append(silhouette_score(X, labels))
                    inertias.append(km.inertia_)

                optimal_idx = np.argmax(scores)
                state["optimal_k"] = k_range[optimal_idx]
                state["k_scores"] = list(zip(k_range, scores, inertias))
                print(f"âœ… Optimal K (clusters): {state['optimal_k']}")
            else:
                state["optimal_k"] = 3  # Default for KNN
                state["k_scores"] = []
                print(f"âœ… Using K (neighbors): {state['optimal_k']}")

        except Exception as e:
            state["error"] = f"Hyperparameter optimization failed: {e}"
            print(f"âŒ {state['error']}")

        return state

    def execute_algorithm_node(self, state: MLState) -> MLState:
        """Node 5: Execute KMeans or KNN"""
        if state.get("error"):
            return state

        try:
            print("ðŸš€ Executing algorithm...")
            X = state["preprocessed_data"]
            k = state["optimal_k"]

            if state["problem_type"] == "clustering":
                # KMeans clustering
                km = KMeans(n_clusters=k, n_init=10, random_state=42)
                labels = km.fit_predict(X)

                state["results"] = {
                    "optimal_k": k,
                    "silhouette_score": silhouette_score(X, labels),
                    "inertia": km.inertia_,
                    "labels": labels,
                    "centers": km.cluster_centers_,
                    "X": X
                }
                print(f"âœ… KMeans completed - Silhouette Score: {state['results']['silhouette_score']:.4f}")

            else:
                # KNN classification
                Xtr, Xte, ytr, yte = train_test_split(
                    X, state["target"], test_size=0.2, random_state=42
                )
                knn = KNeighborsClassifier(n_neighbors=k)
                knn.fit(Xtr, ytr)
                preds = knn.predict(Xte)

                state["results"] = {
                    "k": k,
                    "accuracy": accuracy_score(yte, preds),
                    "precision": precision_score(yte, preds, average='weighted', zero_division=0),
                    "recall": recall_score(yte, preds, average='weighted', zero_division=0),
                    "f1_score": f1_score(yte, preds, average='weighted', zero_division=0),
                    "confusion_matrix": confusion_matrix(yte, preds),
                    "predictions": preds,
                    "actual": yte,
                    "X_test": Xte,
                    "X_train": Xtr,
                    "y_train": ytr
                }
                print(f"âœ… KNN completed - Accuracy: {state['results']['accuracy']:.4f}")

        except Exception as e:
            state["error"] = f"Algorithm execution failed: {e}"
            print(f"âŒ {state['error']}")

        return state

    def run(self, description: str, dataset_path: str) -> MLState:
        """Run the complete workflow"""
        return self.workflow.invoke({
            "description": description,
            "dataset_path": dataset_path,
            "dataset": None,
            "original_dataset": None,
            "problem_type": "",
            "llm_reasoning": "",
            "target_column": "",
            "preprocessed_data": None,
            "target": None,
            "optimal_k": 0,
            "k_scores": [],
            "results": {},
            "error": ""
        })


# =========================
# RICH VISUAL DASHBOARD
# =========================

def create_metrics_summary(state: MLState):
    """Create a beautiful metrics summary visualization"""

    fig = plt.figure(figsize=(16, 10))
    gs = fig.add_gridspec(3, 3, hspace=0.4, wspace=0.3)

    # Title
    fig.suptitle('ðŸ¤– AGENTIC ML WORKFLOW DASHBOARD',
                 fontsize=24, fontweight='bold', color='#2C3E50')

    # ==================
    # LLM ANALYSIS PANEL
    # ==================
    ax_llm = fig.add_subplot(gs[0, :])
    ax_llm.axis('off')

    llm_text = f"""
    â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
    â•‘                          ðŸ§  LLM ANALYSIS SUMMARY                            â•‘
    â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

    ðŸ“Š Problem Type Detected: {state['problem_type'].upper()}
    ðŸŽ¯ Target Column: {state.get('target_column', 'N/A')}
    ðŸ“ Dataset Shape: {state['original_dataset'].shape[0]} rows Ã— {state['original_dataset'].shape[1]} columns
    ðŸ” Features Used: {state['preprocessed_data'].shape[1]}

    ðŸ’¡ LLM Reasoning:
    {state['llm_reasoning']}
    """

    ax_llm.text(0.05, 0.5, llm_text, fontsize=11, family='monospace',
                verticalalignment='center', bbox=dict(boxstyle='round',
                facecolor='#E8F4F8', edgecolor='#3498DB', linewidth=2))

    # ==================
    # ALGORITHM METRICS
    # ==================
    results = state['results']

    if state['problem_type'] == 'clustering':
        # CLUSTERING METRICS

        # Metric Cards
        ax_metrics = fig.add_subplot(gs[1, :])
        ax_metrics.axis('off')

        metrics_text = f"""
        â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
        â•‘                        ðŸ“ˆ CLUSTERING METRICS (K-MEANS)                       â•‘
        â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

        âœ… Optimal Clusters (K): {results['optimal_k']}
        ðŸ“Š Silhouette Score: {results['silhouette_score']:.4f} (Quality: {'Excellent' if results['silhouette_score'] > 0.7 else 'Good' if results['silhouette_score'] > 0.5 else 'Fair'})
        ðŸ“‰ Inertia (Within-cluster sum of squares): {results['inertia']:.2f}
        ðŸŽ¯ Total Samples Clustered: {len(results['labels'])}
        """

        ax_metrics.text(0.05, 0.5, metrics_text, fontsize=12, family='monospace',
                       verticalalignment='center', bbox=dict(boxstyle='round',
                       facecolor='#E8F5E9', edgecolor='#4CAF50', linewidth=2))

        # Cluster Distribution Bar Chart
        ax_bar = fig.add_subplot(gs[2, 0])
        unique, counts = np.unique(results['labels'], return_counts=True)
        colors = plt.cm.Set3(np.linspace(0, 1, len(unique)))
        bars = ax_bar.bar(unique, counts, color=colors, edgecolor='black', linewidth=1.5)
        ax_bar.set_xlabel('Cluster ID', fontsize=12, fontweight='bold')
        ax_bar.set_ylabel('Number of Samples', fontsize=12, fontweight='bold')
        ax_bar.set_title('ðŸ“Š Cluster Distribution', fontsize=14, fontweight='bold', pad=10)
        ax_bar.grid(axis='y', alpha=0.3)

        # Add value labels on bars
        for bar in bars:
            height = bar.get_height()
            ax_bar.text(bar.get_x() + bar.get_width()/2., height,
                       f'{int(height)}',
                       ha='center', va='bottom', fontweight='bold', fontsize=10)

        # Elbow Method Visualization
        ax_elbow = fig.add_subplot(gs[2, 1])
        if state['k_scores']:
            k_vals, silhouette_vals, inertia_vals = zip(*state['k_scores'])
            ax_elbow.plot(k_vals, silhouette_vals, marker='o', linewidth=3,
                         markersize=10, color='#E74C3C', label='Silhouette Score')
            ax_elbow.axvline(results['optimal_k'], color='#27AE60', linestyle='--',
                           linewidth=2, label=f"Optimal K={results['optimal_k']}")
            ax_elbow.set_xlabel('Number of Clusters (K)', fontsize=12, fontweight='bold')
            ax_elbow.set_ylabel('Silhouette Score', fontsize=12, fontweight='bold')
            ax_elbow.set_title('ðŸ“ˆ Elbow Method Analysis', fontsize=14, fontweight='bold', pad=10)
            ax_elbow.legend(fontsize=10)
            ax_elbow.grid(True, alpha=0.3)

        # Cluster Scatter Plot (PCA)
        ax_scatter = fig.add_subplot(gs[2, 2])
        X = results['X']
        labels = results['labels']

        if X.shape[1] > 2:
            pca = PCA(n_components=2)
            X_2d = pca.fit_transform(X)
            xlabel, ylabel = f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)', \
                           f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)'
        else:
            X_2d = X[:, :2]
            xlabel, ylabel = 'Feature 1', 'Feature 2'

        scatter = ax_scatter.scatter(X_2d[:, 0], X_2d[:, 1], c=labels,
                                     cmap='viridis', s=80, alpha=0.7,
                                     edgecolors='black', linewidth=0.5)
        ax_scatter.set_xlabel(xlabel, fontsize=12, fontweight='bold')
        ax_scatter.set_ylabel(ylabel, fontsize=12, fontweight='bold')
        ax_scatter.set_title('ðŸŽ¨ Cluster Visualization', fontsize=14, fontweight='bold', pad=10)
        plt.colorbar(scatter, ax=ax_scatter, label='Cluster ID')
        ax_scatter.grid(True, alpha=0.3)

    else:
        # CLASSIFICATION METRICS

        # Metric Cards
        ax_metrics = fig.add_subplot(gs[1, :])
        ax_metrics.axis('off')

        metrics_text = f"""
        â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
        â•‘                      ðŸŽ¯ CLASSIFICATION METRICS (K-NN)                        â•‘
        â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

        âœ… K (Neighbors): {results['k']}
        ðŸŽ¯ Accuracy: {results['accuracy']:.4f} ({results['accuracy']*100:.2f}%)
        ðŸ“Š Precision: {results['precision']:.4f}
        ðŸ“ˆ Recall: {results['recall']:.4f}
        ðŸ† F1-Score: {results['f1_score']:.4f}
        ðŸ“¦ Test Samples: {len(results['predictions'])}
        """

        ax_metrics.text(0.05, 0.5, metrics_text, fontsize=12, family='monospace',
                       verticalalignment='center', bbox=dict(boxstyle='round',
                       facecolor='#FFF3E0', edgecolor='#FF9800', linewidth=2))

        # Class Distribution
        ax_bar = fig.add_subplot(gs[2, 0])
        unique, counts = np.unique(results['predictions'], return_counts=True)
        colors = plt.cm.Paired(np.linspace(0, 1, len(unique)))
        bars = ax_bar.bar(unique, counts, color=colors, edgecolor='black', linewidth=1.5)
        ax_bar.set_xlabel('Predicted Class', fontsize=12, fontweight='bold')
        ax_bar.set_ylabel('Number of Samples', fontsize=12, fontweight='bold')
        ax_bar.set_title('ðŸ“Š Predicted Class Distribution', fontsize=14, fontweight='bold', pad=10)
        ax_bar.grid(axis='y', alpha=0.3)

        # Add value labels
        for bar in bars:
            height = bar.get_height()
            ax_bar.text(bar.get_x() + bar.get_width()/2., height,
                       f'{int(height)}',
                       ha='center', va='bottom', fontweight='bold', fontsize=10)

        # Most common class highlight
        most_common_idx = np.argmax(counts)
        ax_bar.text(0.5, 0.95, f'ðŸ† Most Common: Class {unique[most_common_idx]}',
                   transform=ax_bar.transAxes, ha='center', va='top',
                   fontsize=11, fontweight='bold',
                   bbox=dict(boxstyle='round', facecolor='gold', alpha=0.7))

        # Confusion Matrix Heatmap
        ax_cm = fig.add_subplot(gs[2, 1])
        cm = results['confusion_matrix']
        sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrRd', cbar_kws={'label': 'Count'},
                   ax=ax_cm, linewidths=1, linecolor='black', square=True)
        ax_cm.set_xlabel('Predicted Class', fontsize=12, fontweight='bold')
        ax_cm.set_ylabel('Actual Class', fontsize=12, fontweight='bold')
        ax_cm.set_title('ðŸ”¥ Confusion Matrix', fontsize=14, fontweight='bold', pad=10)

        # Classification Scatter Plot
        ax_scatter = fig.add_subplot(gs[2, 2])
        X = results['X_test']
        preds = results['predictions']

        if X.shape[1] > 2:
            pca = PCA(n_components=2)
            X_2d = pca.fit_transform(X)
            xlabel, ylabel = f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)', \
                           f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)'
        else:
            X_2d = X[:, :2]
            xlabel, ylabel = 'Feature 1', 'Feature 2'

        scatter = ax_scatter.scatter(X_2d[:, 0], X_2d[:, 1], c=preds,
                                     cmap='coolwarm', s=80, alpha=0.7,
                                     edgecolors='black', linewidth=0.5)
        ax_scatter.set_xlabel(xlabel, fontsize=12, fontweight='bold')
        ax_scatter.set_ylabel(ylabel, fontsize=12, fontweight='bold')
        ax_scatter.set_title('ðŸŽ¨ Classification Visualization', fontsize=14, fontweight='bold', pad=10)
        plt.colorbar(scatter, ax=ax_scatter, label='Predicted Class')
        ax_scatter.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


def print_results_table(state: MLState):
    """Print detailed results in table format"""

    print("\n" + "="*80)
    print("ðŸ“‹ DETAILED RESULTS TABLE")
    print("="*80)

    # Dataset Info Table
    df = state['original_dataset']

    print("\nðŸ“Š DATASET INFORMATION")
    print("-"*80)
    data_info = pd.DataFrame({
        'Metric': ['Total Rows', 'Total Columns', 'Numeric Features', 'Missing Values',
                   'Target Column', 'Problem Type'],
        'Value': [
            df.shape[0],
            df.shape[1],
            len(df.select_dtypes(include=[np.number]).columns),
            df.isnull().sum().sum(),
            state.get('target_column', 'N/A'),
            state['problem_type'].upper()
        ]
    })
    print(data_info.to_string(index=False))

    results = state['results']

    if state['problem_type'] == 'clustering':
        print("\nðŸŽ¯ CLUSTERING RESULTS")
        print("-"*80)

        cluster_info = pd.DataFrame({
            'Cluster ID': np.unique(results['labels']),
            'Sample Count': [np.sum(results['labels'] == i) for i in np.unique(results['labels'])],
            'Percentage': [f"{np.sum(results['labels'] == i)/len(results['labels'])*100:.1f}%"
                          for i in np.unique(results['labels'])]
        })
        print(cluster_info.to_string(index=False))

        print(f"\nðŸ“ˆ Quality Metrics:")
        print(f"   â€¢ Silhouette Score: {results['silhouette_score']:.4f}")
        print(f"   â€¢ Inertia: {results['inertia']:.2f}")
        print(f"   â€¢ Optimal K: {results['optimal_k']}")

    else:
        print("\nðŸŽ¯ CLASSIFICATION RESULTS")
        print("-"*80)

        class_info = pd.DataFrame({
            'Class ID': np.unique(results['predictions']),
            'Predicted Count': [np.sum(results['predictions'] == i) for i in np.unique(results['predictions'])],
            'Percentage': [f"{np.sum(results['predictions'] == i)/len(results['predictions'])*100:.1f}%"
                          for i in np.unique(results['predictions'])]
        })
        print(class_info.to_string(index=False))

        print(f"\nðŸ“Š Performance Metrics:")
        print(f"   â€¢ Accuracy:  {results['accuracy']:.4f} ({results['accuracy']*100:.2f}%)")
        print(f"   â€¢ Precision: {results['precision']:.4f}")
        print(f"   â€¢ Recall:    {results['recall']:.4f}")
        print(f"   â€¢ F1-Score:  {results['f1_score']:.4f}")
        print(f"   â€¢ K Value:   {results['k']}")

    print("\n" + "="*80)


# =========================
# CLI RUNNER
# =========================

def run_ml_pipeline_cli():
    """Run the ML pipeline in CLI mode"""
    print("="*80)
    print("ðŸ¤– AGENTIC ML WORKFLOW â€“ ENHANCED VISUAL DASHBOARD")
    print("   Powered by Mistral 7B + LangGraph + Rich Visualizations")
    print("="*80)

    # Load environment variables from myenv file
    load_dotenv('myenv')

    # Try to get API key from environment variable
    api_key = os.getenv('MISTRAL_API_KEY')

    if api_key:
        print("\nâœ… API Key loaded from myenv file")
    else:
        # Fallback to manual input if not found in .env
        print("\nâš ï¸ MISTRAL_API_KEY not found in myenv file")
        api_key = input("ðŸ”‘ Enter Mistral API Key manually: ").strip()
        if not api_key:
            print("âŒ API key is required")
            return

    # Get Dataset Path
    dataset_path = input("ðŸ“‚ Enter path to CSV dataset: ").strip()
    if not os.path.exists(dataset_path):
        print(f"âŒ Dataset file not found: {dataset_path}")
        return

    # Get Description
    print("\nðŸ“ Describe your ML problem:")
    description = input("> ").strip()
    if not description:
        print("âŒ Description is required")
        return

    print("\n" + "="*80)
    print("ðŸš€ STARTING AGENTIC WORKFLOW...")
    print("="*80 + "\n")

    # Run workflow
    try:
        workflow = MLAgenticWorkflow(api_key)
        state = workflow.run(description, dataset_path)

        if state.get("error"):
            print(f"\nâŒ WORKFLOW FAILED: {state['error']}")
            return

        # Display rich results
        print("\n" + "="*80)
        print("ðŸŽ‰ WORKFLOW COMPLETED SUCCESSFULLY!")
        print("="*80)

        # Print detailed table
        print_results_table(state)

        # Show visual dashboard
        print("\nðŸ“Š Generating Rich Visual Dashboard...")
        create_metrics_summary(state)

    except Exception as e:
        print(f"\nâŒ FATAL ERROR: {e}")
        import traceback
        traceback.print_exc()


# =========================
# ENTRY POINT
# =========================

if __name__ == "__main__":
    run_ml_pipeline_cli()
````
